```python
import matplotlib.pyplot as plt
import random

score = 0

xmin = -20
xmax = 20
ymin = -20
ymax = 20

fig, ax = plt.subplots()

while True:
    xpoint = random.randint(xmin, xmax)
    ypoint = random.randint(ymin, ymax)
    x = [xpoint]
    y = [ypoint]
    plt.axis([xmin,xmax,ymin,ymax])
    plt.plot([xmin,xmax],[0,0],'b')
    plt.plot([0,0],[ymin,ymax], 'b')
    plt.plot(x, y, 'ro')
    print(" ")
    plt.grid()
    plt.show()
    guess = input("Enter 'exit' to exit or enter the coordinates of the red point point: \n")
    if guess.lower() == 'exit':
        break
    guess_array = guess.split(",")
    xguess = int(guess_array[0])
    yguess = int(guess_array[1])
    if xguess == xpoint and yguess == ypoint:
        score = score + 1
        print('Correct')
    else:
        print('Incorrect')

print("Final score: ", score)
```

```python
import random

score = 0

# structure: string, function giving correct answer, function evaluating condition, number of values needed
one_step = [
    ('If you have {} candies and you receive {} more, how many candies do you have in total?',
     lambda value1, value2: value1 + value2,
     lambda value1, value2: True,
     2),
    ('If you have ${} and you spend ${} on a toy, how much money do you have left?',
     lambda value1, value2: value1 - value2,
     lambda value1, value2: value1 >= value2,
     2),
    ('If you have {} boxes, and each box contains {} pencils, how many pencils do you have in total?',
     lambda value1, value2: value1 * value2,
     lambda value1, value2: value1 >= value2,
     2),
    ('If you have {} cookies, and you want to share them equally among {} friends, how many cookies will each friend get?',
     lambda value1, value2: value1 / value2,
     lambda value1, value2: (value1 % value2) == 0,
     2),
    ('You have {} marbles, and you want to put them into {} bags with an equal number of marbles in each bag. How many marbles will be in each bag?',
     lambda value1, value2: value1 / value2,
     lambda value1, value2: (value1 % value2) == 0,
     2)
]

two_step = [
    ('You want to buy {} books, and each book costs ${}. You also want to buy {} pencils, and each pencil costs ${}. How much money will you spend in total?',
     lambda value1, value2, value3, value4: value1 * value2 + value3 * value4,
     lambda value1, value2, value3, value4: True,
     4),
    ('You have ${}, and you want to buy a toy that costs ${}. After buying the toy, you want to save half of the remaining money. How much money will you have left?',
     lambda value1, value2: (value1 - value2) / 2,
     lambda value1, value2: value1 >= value2,
     2),
    ('You have a garden with {} rows of flowers, and each row has {} flowers. You decide to pick {} flowers to put in a vase. How many flowers are left in the garden?',
     lambda value1, value2, value3: value1 * value2 - value3,
     lambda value1, value2, value3: value1 * value2 >= value3,
     3),
    ('You have {} cookies, and you want to share them equally among you and {} friends. After sharing the cookies, you decide to bake {} more. How many cookies do you have in total now?',
     lambda value1, value2, value3: value1 / (value2 + 1) + value3,
     lambda value1, value2, value3: (value1 % (value2 + 1) == 0 and value1 >= (value2 + 1)),
     3),
    ('You have a pizza with {} slices, and you eat {} slices. Then, your friend gives you {} more slices. How many pizza slices do you have now?',
     lambda value1, value2, value3: value1 - value2 + value3,
     lambda value1, value2, value3: value1 >= value2,
     3)
]

try:
    while True:
        problems = random.choice([one_step, two_step])
        problem = random.choice(problems)
        sentence = problem[0]
        correct_answer = problem[1]
        condition = problem[2]
        n_values = problem[3]

        values = [random.randint(1, 100) for _ in range(n_values)]
        while not condition(*values):
            values = [random.randint(1, 100) for _ in range(n_values)]

        answer = input(sentence.format(*values) + ' ')
        if int(answer) == correct_answer(*values):
            print('Correct')
            score += 1
        else:
            print('Incorrect')
except KeyboardInterrupt:
    print("Final score: ", score)
```
