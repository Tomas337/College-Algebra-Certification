```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import io
import numpy as np

def get_table():
    print('Upload a csv file')
    upload_method = input('Input "1" to upload from device, "2" to input URL, "3" to use the default hardcoded file. ')

    table = None

    if upload_method == "1":
        uploaded = files.upload()
        file_name = next(iter(uploaded))
        table = pd.read_csv(io.BytesIO(uploaded[file_name]))
    elif upload_method == "2":
        try:
            url = input('Input URL: ')
            table = pd.read_csv(url)
        except:
            print('Invalid URL\n')
            return get_table()
    elif upload_method == "3":
        table = pd.read_csv('./sample_data/california_housing_test.csv')
    else:
        print('Incorrect upload method\n')
        return get_table()

    return table

def get_ndarrays(df):
    cols = input('Input 1 or 2 column names separeted by ",": ')
    cols = cols.split(',')

    output = []
    col_names = list(df.columns)

    for name in cols:
        if name in col_names:
            output.append(df[name].to_numpy())
        else:
            print('Invalid column name\n')
            return get_ndarrays(df)
    return output, cols

def get_display_type():
    display_type = input('Display data as [s]catter plot or [l]ine? ').lower()
    if display_type != 's' and display_type != 'l':
        print('Incorrect input\n')
        return get_display_type()
    return display_type

try:
    table = get_table()
    df = pd.DataFrame(table)

    print(df.head(2))

    nd_arrays, columns = get_ndarrays(df)

    y = nd_arrays[0]
    if len(nd_arrays) == 1:
        x = np.arange(len(y))
    else:
        x = nd_arrays[1]
        sorted_indices = np.argsort(x)
        x = x[sorted_indices]
        y = y[sorted_indices]
        plt.xlabel(columns[1])
    plt.ylabel(columns[0])

    display_type = get_display_type()
    if display_type == 'l':
        plt.plot(x, y, color='b')
    else:
        plt.scatter(x, y, color='b', marker='o')
    plt.show()
except KeyboardInterrupt:
    print('Ending\n')
except:
    print('Something went wrong\n')
```
